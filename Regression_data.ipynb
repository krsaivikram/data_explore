Why is drop_first=True used in one-hot encoding, and what does a row of all zeros in the new dummy columns represent?

In one-hot encoding, drop_first=True is used to avoid multicollinearity, also known as the dummy variable trap, where one dummy variable can be predicted from others. By dropping one category, it becomes the reference group, and the model uses it as a baseline for comparison. A row with all zeros in the dummy columns represents this dropped reference category, meaning the observation belongs to that category which was intentionally removed during encoding.
